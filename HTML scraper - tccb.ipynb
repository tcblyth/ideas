```python
import requests 
from bs4 import BeautifulSoup 

URL = "https://t.me/s/netflix"
r = requests.get(URL) 

soup = BeautifulSoup(r.content, 'html5lib') 

results = list(map(lambda x: x.get_text(), soup.find_all("div", {"class": "tgme_widget_message_text js-message_text"})))

list(results)
```
